```python
import rasterio
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
Urban= rasterio.open("/Users/arngolo/Downloads/RESEARCH_Angolo/Integrating-GIS-RemoteSensing-ML-for-monitoring-urban-expansion/Parallel_Coordinate_Sample_Data/MOSAIC_Luanda_2018_L8/NDBI_MEI_VIGS_DBI_QzCal_VV_VH_Parallel_Coordinate/Urban_parallel_coord.tif")
Bareland= rasterio.open("/Users/arngolo/Downloads/RESEARCH_Angolo/Integrating-GIS-RemoteSensing-ML-for-monitoring-urban-expansion/Parallel_Coordinate_Sample_Data/MOSAIC_Luanda_2018_L8/NDBI_MEI_VIGS_DBI_QzCal_VV_VH_Parallel_Coordinate/Bareland_parallel_coord.tif")
Water= rasterio.open("/Users/arngolo/Downloads/RESEARCH_Angolo/Integrating-GIS-RemoteSensing-ML-for-monitoring-urban-expansion/Parallel_Coordinate_Sample_Data/MOSAIC_Luanda_2018_L8/NDBI_MEI_VIGS_DBI_QzCal_VV_VH_Parallel_Coordinate/Water_parallel_coord.tif")
Vegetation= rasterio.open("/Users/arngolo/Downloads/RESEARCH_Angolo/Integrating-GIS-RemoteSensing-ML-for-monitoring-urban-expansion/Parallel_Coordinate_Sample_Data/MOSAIC_Luanda_2018_L8/NDBI_MEI_VIGS_DBI_QzCal_VV_VH_Parallel_Coordinate/Vegetation_parallel_coord.tif")
```

```python
Urban_NDBI=np.ma.masked_outside(Urban.read(1), -1,1)
Urban_MEI=np.ma.masked_outside(Urban.read(2), -1,1)
Urban_VIGS=np.ma.masked_outside(Urban.read(3), -1,1)
Urban_DBI=np.ma.masked_outside(Urban.read(4), -1,1)
Urban_QzCal=np.ma.masked_outside(Urban.read(5), -1,1)
Urban_VV=np.ma.masked_outside(Urban.read(6), -1,1)
Urban_VH=np.ma.masked_outside(Urban.read(7), -1,1)
#######
Bareland_NDBI=np.ma.masked_outside(Bareland.read(1), -1,1)
Bareland_MEI=np.ma.masked_outside(Bareland.read(2), -1,1)
Bareland_VIGS=np.ma.masked_outside(Bareland.read(3), -1,1)
Bareland_DBI=np.ma.masked_outside(Bareland.read(4), -1,1)
Bareland_QzCal=np.ma.masked_outside(Bareland.read(5), -1,1)
Bareland_VV=np.ma.masked_outside(Bareland.read(6), -1,1)
Bareland_VH=np.ma.masked_outside(Bareland.read(7), -1,1)
#######
Water_NDBI=np.ma.masked_outside(Water.read(1), -1,1)
Water_MEI=np.ma.masked_outside(Water.read(2), -1,1)
Water_VIGS=np.ma.masked_outside(Water.read(3), -1,1)
Water_DBI=np.ma.masked_outside(Water.read(4), -1,1)
Water_QzCal=np.ma.masked_outside(Water.read(5), -1,1)
Water_VV=np.ma.masked_outside(Water.read(6), -1,1)
Water_VH=np.ma.masked_outside(Water.read(7), -1,1)
#######
Vegetation_NDBI=np.ma.masked_outside(Vegetation.read(1), -1,1)
Vegetation_MEI=np.ma.masked_outside(Vegetation.read(2), -1,1)
Vegetation_VIGS=np.ma.masked_outside(Vegetation.read(3), -1,1)
Vegetation_DBI=np.ma.masked_outside(Vegetation.read(4), -1,1)
Vegetation_QzCal=np.ma.masked_outside(Vegetation.read(5), -1,1)
Vegetation_VV=np.ma.masked_outside(Vegetation.read(6), -1,1)
Vegetation_VH=np.ma.masked_outside(Vegetation.read(7), -1,1)
```

```python
####
URban = pd.DataFrame({"NDBI":Urban_NDBI[Urban_NDBI==Urban_NDBI],"MEI":Urban_MEI[Urban_MEI==Urban_MEI],"VIGS":Urban_VIGS[Urban_VIGS==Urban_VIGS], "DBI":Urban_DBI[Urban_DBI==Urban_DBI], "QzCal":Urban_QzCal[Urban_QzCal==Urban_QzCal], "VV":Urban_VV[Urban_VV==Urban_VV], "VH":Urban_VH[Urban_VH==Urban_VH]})
####
BAreland = pd.DataFrame({"NDBI":Bareland_NDBI[Bareland_NDBI==Bareland_NDBI],"MEI":Bareland_MEI[Bareland_MEI==Bareland_MEI],"VIGS":Bareland_VIGS[Bareland_VIGS==Bareland_VIGS], "DBI":Bareland_DBI[Bareland_DBI==Bareland_DBI], "QzCal":Bareland_QzCal[Bareland_QzCal==Bareland_QzCal], "VV":Bareland_VV[Bareland_VV==Bareland_VV], "VH":Bareland_VH[Bareland_VH==Bareland_VH]})
####
WAter = pd.DataFrame({"NDBI":Water_NDBI[Water_NDBI==Water_NDBI],"MEI":Water_MEI[Water_MEI==Water_MEI],"VIGS":Water_VIGS[Water_VIGS==Water_VIGS], "DBI":Water_DBI[Water_DBI==Water_DBI], "QzCal":Water_QzCal[Water_QzCal==Water_QzCal], "VV":Water_VV[Water_VV==Water_VV], "VH":Water_VH[Water_VH==Water_VH]})
####
VEgetation = pd.DataFrame({"NDBI":Vegetation_NDBI[Vegetation_NDBI==Vegetation_NDBI],"MEI":Vegetation_MEI[Vegetation_MEI==Vegetation_MEI],"VIGS":Vegetation_VIGS[Vegetation_VIGS==Vegetation_VIGS], "DBI":Vegetation_DBI[Vegetation_DBI==Vegetation_DBI], "QzCal":Vegetation_QzCal[Vegetation_QzCal==Vegetation_QzCal], "VV":Vegetation_VV[Vegetation_VV==Vegetation_VV], "VH":Vegetation_VH[Vegetation_VH==Vegetation_VH]})
URBan=URban.dropna()
BAReland=BAreland.dropna()
WATer=WAter.dropna()
VEGetation=VEgetation.dropna()
URBan["Label"]= "Urban"
BAReland["Label"]= "Bareland"
WATer["Label"]= "Water"
VEGetation["Label"]= "Vegetation"
URBan.head()
```
